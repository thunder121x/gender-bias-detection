```python
import csv
from datetime import datetime
from yt_dlp import YoutubeDL

video_url = "https://www.youtube.com/watch?v=bQoVZe4Ohbo"

options = {
    'skip_download': True,
    'extract_flat': False,
    'quiet': True,
    'force_generic_extractor': False,
    'extractor_args': {'youtube': {'skip': ['dash']}},
    'noplaylist': True,
    'writeinfojson': True,
    'getcomments': True,
    'dump_single_json': True
}

with YoutubeDL(options) as ydl:
    result = ydl.extract_info(video_url, download=False)
    comments = result.get("comments", [])

# CSV header
fields = [
    "id", "platform", "platform_type", "url",
    "content_type", "timestamp", "scraper_module",
    "text", "raw_text", "scrape_date"
]

rows = []
scrape_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

for c in comments:
    row = {
        "id": c.get("id"),
        "platform": "youtube",
        "platform_type": "video",
        "url": video_url,
        "content_type": "comment",
        "timestamp": datetime.utcfromtimestamp(c.get("timestamp", 0)).strftime("%Y-%m-%d %H:%M:%S") if c.get("timestamp") else "",
        "scraper_module": "yt-dlp",
        "text": c.get("text", "").strip(),
        "raw_text": c.get("text", ""),  # keep as-is
        "scrape_date": scrape_date
    }
    rows.append(row)

# Write to CSV
with open("youtube_comments.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fields)
    writer.writeheader()
    writer.writerows(rows)
```

```python
import csv
from datetime import datetime
from yt_dlp import YoutubeDL

video_url = "https://www.tiktok.com/@nissy_0122/video/7546438874688523538"

options = {
    'skip_download': True,
    'extract_flat': False,
    'quiet': True,
    'force_generic_extractor': False,
    'extractor_args': {'tiktok': {'comments': ['1']}},  # 👈 enable comment extraction
    'noplaylist': True,
    'writeinfojson': True,
    'getcomments': True,
    'dump_single_json': True,
}

with YoutubeDL(options) as ydl:
    result = ydl.extract_info(video_url, download=False)

# Extract comments (if available)
comments = result.get("comments", [])
print(f"✅ Total comments: {len(comments)}")

# Save to CSV
if comments:
    filename = f"tiktok_comments_{result['id']}.csv"
    with open(filename, mode="w", encoding="utf-8-sig", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["id", "platform", "platform_type", "url", "content_type", "timestamp", "author", "content"])
        for i, c in enumerate(comments):
            writer.writerow([
                c.get("id", f"comment_{i}"),
                "tiktok",
                "video",
                video_url,
                "comment",
                datetime.fromtimestamp(c.get("timestamp", 0)).isoformat() if c.get("timestamp") else "",
                c.get("author", {}).get("nickname", "unknown"),
                c.get("text", ""),
            ])
    print(f"💾 Saved to {filename}")
else:
    print("⚠️ No comments found — TikTok may have limited public comment access.")
```

```python
import asyncio
from playwright.async_api import async_playwright

async def check_playwright():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)
        page = await browser.new_page()
        await page.goto("https://www.tiktok.com")
        print("✅ Chromium launched and opened TikTok successfully!")
        await browser.close()

await check_playwright()
```

```python
import asyncio
import csv
from TikTokApi import TikTokApi

video_url = "https://www.tiktok.com/@nissy_0122/video/7546438874688523538"
video_id = video_url.split("/")[-1]
output_file = "tiktok_comments.csv"

async def scrape_tiktok_comments():
    async with TikTokApi() as api:
        # Step 1. Create Playwright browser sessions
        await api.create_sessions(
            num_sessions=1,       # one browser tab is enough
            headless=False,      # 👈 Visible browser window
            browser="chromium",   # or "webkit" (less bot detection)
            sleep_after=2
        )

        # Step 2. Get video and fetch comments
        video = api.video(id=video_id)
        comments = []
        async for comment in video.comments(count=100):
            comments.append(comment)

        print(f"✅ Got {len(comments)} comments")

        # Step 3. Save to CSV
        with open(output_file, "w", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["comment_id", "username", "text", "create_time"])
            for c in comments:
                writer.writerow([
                    c.id,
                    getattr(c.author, "username", None),
                    c.text,
                    c.create_time
                ])

        print(f"💾 Saved to {output_file}")

        # Step 4. Close sessions
        await api.close_sessions()
        await api.stop_playwright()

import nest_asyncio
import asyncio

nest_asyncio.apply()
await scrape_tiktok_comments()
```

```python
import asyncio
import csv
from TikTokApi import TikTokApi
import nest_asyncio

nest_asyncio.apply()

video_url = "https://www.tiktok.com/@nissy_0122/video/7546438874688523538"
video_id = video_url.split("/")[-1]
output_file = "tiktok_comments.csv"

async def scrape_tiktok_comments():
    async with TikTokApi() as api:
        # Step 1. Create Playwright browser session
        await api.create_sessions(
            num_sessions=1,
            headless=False,
            browser="chromium",
            sleep_after=2
        )

        # Step 2. Fetch comments
        video = api.video(id=video_id)
        comments = []
        async for c in video.comments(count=100):
            comments.append(c)

        print(f"✅ Got {len(comments)} comments")

        # Step 3. Save safely to CSV
        with open(output_file, "w", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["comment_id", "username", "text", "create_time"])
            for c in comments:
                create_time = getattr(c, "create_time", "")
                username = getattr(c.author, "username", None)
                writer.writerow([c.id, username, c.text, create_time])

        print(f"💾 Saved to {output_file}")

        # Step 4. Close sessions
        await api.close_sessions()
        await api.stop_playwright()

await scrape_tiktok_comments()
```
