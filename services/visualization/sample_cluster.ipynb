```python
import pandas as pd

# -----------------------------
# 1) Paths to your topic files
# -----------------------------
files = {
    "bge-m3": "assets/result_clustering/my_thai_topics_results_BAAI_bge-m3_cv_1_2_config_u_scan.csv",
    "wangchan-simcse": "assets/result_clustering/my_thai_topics_results_new5558_simcse-model-wangchanberta-base-att-spm-uncased_cv_1_2_config_u_scan.csv",
    "nomic-v2": "assets/result_clustering/my_thai_topics_results_nomic-ai_nomic-embed-text-v2-moe_cv_1_2_config_u_scan.csv",
}
files_def = {
    "bge-m3": "assets/result_clustering/my_topic_definitions_BAAI_bge-m3_cv_1_2_config_u_scan.csv",
    "wangchan-simcse": "assets/result_clustering/my_topic_definitions_new5558_simcse-model-wangchanberta-base-att-spm-uncased_cv_1_2_config_u_scan.csv",
    "nomic-v2": "assets/result_clustering/my_topic_definitions_nomic-ai_nomic-embed-text-v2-moe_cv_1_2_config_u_scan.csv",
}

# -----------------------------
# 2) Topic numbers of interest
# -----------------------------
topics_of_interest = {
    "bge-m3":    [0, 3, 87, 104],
    "wangchan-simcse": [0, 5, 40],
    "nomic-v2": [0, 116, 124, 67],
}
```

```python
# -----------------------------
# 3) Load data and join topic definitions
# -----------------------------
samples = {}

for model in files:
    df = pd.read_csv(files[model])
    df_def = pd.read_csv(files_def[model])

    # Normalize column names for consistency
    df_def.rename(
        columns={"Topic": "topic", "Name": "topic_name", "Representation": "representation"},
        inplace=True
    )

    samples[model] = {}

    for t in topics_of_interest[model]:

        # get topic definition row
        topic_info = df_def[df_def["topic"] == t]

        if topic_info.empty:
            topic_name = "(no name)"
            representation = "(no representation)"
            representative_docs = "(no docs)"
        else:
            topic_name = topic_info["topic_name"].values[0]
            representation = topic_info["representation"].values[0]
            representative_docs = topic_info["Representative_Docs"].values[0]

        # extract samples
        df_topic = df[df["topic"] == t]
        df_sample = df_topic.sample(n=min(5, len(df_topic)), random_state=42)

        samples[model][t] = {
            "topic_name": topic_name,
            "representation": representation,
            "representative_docs": representative_docs,
            "rows": df_sample
        }
        
```

```python
# -----------------------------
# 4) Print nicely
# -----------------------------
for model in samples:
    print("="*100)
    print(f"MODEL: {model}")
    print("="*100)

    for topic_no, content in samples[model].items():
        print("\n------------------------------------------------------------")
        print(f"TOPIC {topic_no}")
        print("------------------------------------------------------------")
        print(f"Name: {content['topic_name']}")
        print(f"Representation: {content['representation']}")
        print(f"Representative Docs: {content['representative_docs'][:200]}...")
        print("\nSample Documents:")
        print(content["rows"][["topic", "text"]].to_string(index=False))
        print("\n")
        
```
