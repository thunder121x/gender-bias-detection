```python
import pandas as pd
import glob
import os

# Define source and destination
src_dir = "./outputs/pseudo_v2"
dest_dir = "./outputs/pseudo_v3"

# Create the destination directory if it doesn't exist
os.makedirs(dest_dir, exist_ok=True)

# Find all CSV files in the source folder
csv_files = glob.glob(os.path.join(src_dir, "*.csv"))

if not csv_files:
    print(f"No CSV files found in {src_dir}")
else:
    for file_path in csv_files:
        try:
            df = pd.read_csv(file_path)
            file_name = os.path.basename(file_path)

            if "text" in df.columns:
                # Deduplicate
                df_cleaned = df.drop_duplicates(subset=["text"], keep="first")

                # Define new save path
                save_path = os.path.join(dest_dir, file_name)

                # Save to v3
                df_cleaned.to_csv(save_path, index=False)

                removed_count = len(df) - len(df_cleaned)
                print(f"Saved {file_name} to v3: Removed {removed_count} duplicates.")
            else:
                print(f"Skipped {file_name}: 'text' column missing.")

        except Exception as e:
            print(f"Error processing {file_path}: {e}")

    print(f"\nAll files have been processed and saved to: {dest_dir}")
```

```python
import pandas as pd
import glob
import os

# Define source and the new specific column destination
src_dir = "./outputs/pseudo_v3"
dest_dir = "./outputs/pseudo_v3_extracted"

# Ensure the destination directory exists
os.makedirs(dest_dir, exist_ok=True)

# Find all CSV files in the v3 folder
csv_files = glob.glob(os.path.join(src_dir, "*.csv"))

if not csv_files:
    print(f"No CSV files found in {src_dir}. Did you run the deduplication step first?")
else:
    for file_path in csv_files:
        try:
            # Read the deduplicated CSV
            df = pd.read_csv(file_path)
            file_name = os.path.basename(file_path)

            # Check if both required columns exist
            required_cols = ["id", "text", "tokens"]
            if all(col in df.columns for col in required_cols):
                # Select only the specific columns
                df_subset = df[required_cols]

                # Define save path
                save_path = os.path.join(dest_dir, file_name)

                # Save the slimmed-down CSV
                df_subset.to_csv(save_path, index=False)
                print(f"Extracted 'text' and 'tokens' from {file_name} -> {dest_dir}")
            else:
                missing = [c for c in required_cols if c not in df.columns]
                print(f"Skipped {file_name}: Missing columns {missing}")

        except Exception as e:
            print(f"Error processing {file_path}: {e}")

    print(f"\nExtraction complete. Files are located in: {dest_dir}")
```
