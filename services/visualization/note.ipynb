```python
from pythainlp.corpus import thai_stopwords

# thai_stopwords() returns a set already
stopwords = thai_stopwords()

# convert to sorted list if needed
stopwords_list = sorted(stopwords)

# print(stopwords_list)
```

```python
text = " ".join(df["tokens"].explode().tolist())

wc = WordCloud(
    stopwords=stopwords_list,
    width=2000,
    height=1000,
    background_color="white",
    font_path=FONT,
    regexp=r"[ก-๙]+",
    collocations=False,
    prefer_horizontal=1,
    max_words=50, 
    colormap='tab20c',
).generate(text)

plt.figure(figsize=(14, 7))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title(f"Word Cloud Social Media", fontsize=24)

# ✔ บันทึกไฟล์แทน show
save_path = os.path.join(output_dir, f"wordcloud_social_media.png")
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.close()

print(f"Saved: {save_path}")
```
