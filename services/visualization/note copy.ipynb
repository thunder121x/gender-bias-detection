```python

# -----------------------------------------
# 1) Clean raw timestamp strings
# -----------------------------------------
df['timestamp'] = df['timestamp'].astype(str).str.strip()

# Replace blank / invalid strings with NA
df['timestamp'] = df['timestamp'].replace(
    ["", " ", "None", "none", "nan", "NaN", "NULL", "null"],
    pd.NA
)

# -----------------------------------------
# 2) Convert to datetime (robust parser)
# Handles both:
#   - "2019-06-15T13:06:53.000Z"
#   - "2022-12-05 00:00:00"
# -----------------------------------------
df['timestamp'] = pd.to_datetime(
    df['timestamp'],
    errors='coerce',   # avoid crashing
    utc=True           # normalize all formats to UTC
)

# -----------------------------------------
# 3) Optional fallback: use scrape_date
# If tweet had missing timestamp but scraper logged scrape_date
# -----------------------------------------
df['scrape_date'] = pd.to_datetime(df['scrape_date'], errors='coerce', utc=True)

df['timestamp'] = df['timestamp'].fillna(df['scrape_date'])

# -----------------------------------------
# 4) Diagnostics
# -----------------------------------------
print("Total rows:", len(df))
print("Unparsed timestamps:", df['timestamp'].isna().sum())

if df['timestamp'].isna().sum() > 0:
    print("\nExamples of unparsed timestamps:")
    print(df[df['timestamp'].isna()][['timestamp', 'scrape_date']].head(10))

# -----------------------------------------
# 5) Create month column (SAFE now)
# -----------------------------------------
df['month'] = df['timestamp'].dt.to_period('M')

print("\nFinal timestamp dtype:", df['timestamp'].dtype)
print(df[['timestamp', 'month']].head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import matplotlib.font_manager as fm

# ---------------------------------------------------
# Optional: Set Times New Roman globally (fallback to serif)
# ---------------------------------------------------
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["axes.edgecolor"] = "#BBBBBB"
plt.rcParams["axes.labelcolor"] = "#252525"
plt.rcParams["xtick.color"] = "#252525"
plt.rcParams["ytick.color"] = "#252525"
plt.rcParams["text.color"] = "#252525"

# ---------------------------------------------------
# Soft Pastel Palette
# ---------------------------------------------------
PASTEL = {
    "overall": "#a8b1d0",   # muted lavender
    "twitter": "#9ec5fe",   # soft baby blue
    "tiktok":  "#b7e4c7",   # pastel green
    "youtube": "#fbc4ab",   # soft peach/coral
}

# ---------------------------------------------------
# Minimal KDE Function
# ---------------------------------------------------
def kde_curve(data, label, color):
    if len(data) < 2:
        print(f"[WARN] Not enough data for KDE:", label)
        return

    kde = gaussian_kde(data)
    x_vals = np.linspace(np.min(data), np.max(data), 450)
    plt.plot(
        x_vals, kde(x_vals),
        label=label,
        color=color,
        linewidth=2.5,
        alpha=0.9
    )

# ---------------------------------------------------
# Extract token lengths
# ---------------------------------------------------
df["token_len"] = df["tokens"].apply(len)

overall = df["token_len"].values
tw = df[df["platform"] == "twitter"]["token_len"].values
tt = df[df["platform"] == "tiktok"]["token_len"].values
yt = df[df["platform"] == "youtube"]["token_len"].values

# ---------------------------------------------------
# Plot
# ---------------------------------------------------
plt.figure(figsize=(11, 5))

kde_curve(overall, "Overall", PASTEL["overall"])
kde_curve(tw, "Twitter", PASTEL["twitter"])
kde_curve(tt, "TikTok", PASTEL["tiktok"])
kde_curve(yt, "YouTube", PASTEL["youtube"])

plt.title("Token Length Distribution Across Platforms", fontsize=18, pad=12)
plt.xlabel("Token Count", fontsize=14)
plt.ylabel("Density", fontsize=14)

# Clean Legend
plt.legend(frameon=False, fontsize=12)

# Remove grid
plt.grid(False)

# Light axis lines
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

# Keep bottom/left but soften
plt.gca().spines["bottom"].set_linewidth(0.8)
plt.gca().spines["left"].set_linewidth(0.8)

plt.tight_layout()
plt.show()
```

```python
import ast
from collections import Counter
import pandas as pd

# ---------------------------------
# 1. เตรียม tokens
# ---------------------------------
# ถ้า df["tokens"] ยังไม่ tokenize มาก่อน ให้ tokenize:
# df["tokens"] = df["text"].apply(thai_tokenize)

# Flatten token list
df["tokens"] = df["tokens"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
all_tokens = [t for row in df["tokens"] for t in row]

freq = Counter(all_tokens)

# ---------------------------------
# 2. ดูสรุปพื้นฐาน
# ---------------------------------
print("🔹 จำนวน token ทั้งหมด:", len(all_tokens))
print("🔹 จำนวน unique tokens:", len(freq))
print()

# ---------------------------------
# 3. แสดง top 50 tokens ที่เจอบ่อยสุด
# ---------------------------------
print("🔥 Top 50 Most Frequent Tokens")
print(pd.DataFrame(freq.most_common(50), columns=["token", "count"]))
print()

# ---------------------------------
# 4. แสดง rare tokens ที่พบ < 5 ครั้งแรกสุด
# ---------------------------------
rare_tokens = [(t, c) for t, c in freq.items() if c < 5]
rare_tokens_sorted = sorted(rare_tokens, key=lambda x: x[1])

print("⚠️ Rare Tokens (<5 occurrences) — First 50")
print(pd.DataFrame(rare_tokens_sorted[:50], columns=["token", "count"]))
print()

# ---------------------------------
# 5. แสดง tokens ที่จะถูกลบตาม rule ของคุณ
# Rule: rare (<5) + all chars identical + length > 3
# ---------------------------------
to_delete = [
    t for t, c in freq.items()
    if c < 5 and len(set(t)) == 1 and len(t) > 3
]

print("🗑️ Tokens that WOULD be removed by your rule")
print(pd.DataFrame({"token": to_delete, "count": [freq[t] for t in to_delete]}).head(50))
print("\nTotal tokens to remove:", len(to_delete))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# ---------------------------------------------------
# Global styling (Times New Roman + soft greys)
# ---------------------------------------------------
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["axes.labelcolor"] = "#252525"
plt.rcParams["xtick.color"] = "#252525"
plt.rcParams["ytick.color"] = "#252525"
plt.rcParams["text.color"] = "#252525"

# ---------------------------------------------------
# Pastel colors from earlier
# ---------------------------------------------------
PASTEL_LABEL = {
    "non_toxic": "#b7e4c7",   # mint green
    "toxic":     "#f5b5b5",   # soft pink
    "safe":      "#c7e8ff",   # baby blue
    "unsafe":    "#ffd6a5",   # pastel apricot
}

# ---------------------------------------------------
# Build a soft blended heatmap colormap
# ---------------------------------------------------
colors = [
    PASTEL_LABEL["toxic"],
    PASTEL_LABEL["unsafe"],
    PASTEL_LABEL["safe"],
    PASTEL_LABEL["non_toxic"]
]
pastel_cmap = LinearSegmentedColormap.from_list("pastel_blend", colors, N=256)

# ---------------------------------------------------
# Plot the heatmap
# counts is a 2×2 matrix like:
# [[Toxic-Safe, Toxic-Unsafe],
#  [NonToxic-Safe, NonToxic-Unsafe]]
# ---------------------------------------------------
plt.figure(figsize=(8, 6))

ax = sns.heatmap(
    counts,
    annot=True,
    fmt="d",
    cmap=pastel_cmap,
    linewidths=0.5,
    linecolor="#EEEEEE",
    cbar=False,
    xticklabels=["Safe", "Unsafe"],
    yticklabels=["Toxic", "Non-Toxic"]
)

plt.title("Toxicity × Safety Classification Matrix (4-way Intersection)", fontsize=16)
plt.xlabel("Safety Model", fontsize=14)
plt.ylabel("Toxicity Model", fontsize=14)

plt.tight_layout()
plt.show()
```
